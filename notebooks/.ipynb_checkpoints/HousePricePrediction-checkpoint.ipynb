```python
name = 'housing.csv'
```

```python
def get_dataset_path(name):
    file_name = name
    search_dir = os.getcwd()
    for roots, dirs, files in os.walk(search_dir):
        if file_name in files:
            dataset_path = os.path.join(roots, file_name)
            return dataset_path
        else:
            continue
    return False
```

```python
dataset_path = get_dataset_path(name)
dataset_path
```

### Loading Dataset

```python
df = pd.read_csv(dataset_path)
```

```python
df.dropna(inplace=True)
```

### Data Preprocessing

```python
df.ocean_proximity.unique()
```

```python
df = pd.get_dummies(df, columns=['ocean_proximity'], dtype=int)
```

```python
df['rooms_per_house'] = df['total_rooms'] / df['households']
df['bedrooms_per_house'] = df['total_bedrooms'] / df['households']
```

```python
df.drop(['total_rooms', 'total_bedrooms', 'population', 'households'], axis=1, inplace=True)
```

```python
house_value = df.pop('median_house_value')
df.insert(len(df.columns), 'median_house_value', house_value)
```

## Model Training

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['median_house_value'], axis=1)
Y = df['median_house_value']
```

```python
from sklearn.feature_selection import mutual_info_regression
mi_scores = mutual_info_regression(X, Y, random_state=42, n_jobs=-1)
mi_df = pd.DataFrame({'Features': X.columns, 'Scores': mi_scores})
```

```python
mi_df = mi_df.sort_values(by='Scores', ascending=False)
```

```python
plt.figure(figsize=(16, 10))
plt.title('Mutual Info Regression Scores')
sns.barplot(mi_df, x=mi_df['Scores'], y=mi_df['Features'])

plt.savefig('../reports/mi_scores.png')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.75, random_state = 42)
```

### Data Scaling/Transforming

```python
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
fig, _ = plt.subplots(3,3, figsize=(16,16))
fig.suptitle('After Scaling for Outliers')

for i in range(2, len(df.columns)-1):
    plt.subplot(3,3, i-1)
    sns.boxplot(x_train_scaled[:, i])
    plt.title(f'Column - {i}')

plt.savefig('../reports/boxplot2.png')
```

### Importing Models for training

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor 
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
```

### RandomizedSearchCV to find the best performing features and ML algorithm

```python
pipeline = Pipeline([
    ('feature_selection', SelectKBest(score_func=mutual_info_regression)),
     ('regressor', RandomForestRegressor())
])
```

```python
par_grid = [
    {'regressor': [RandomForestRegressor(random_state=42)], 
    'regressor__n_estimators': [10, 50, 100, 200],
     'feature_selection__k': [3, 5, 7, 9]
    },
    {'regressor': [KNeighborsRegressor()],
    'regressor__n_neighbors': [5, 10, 15], 
    'regressor__weights': ['uniform', 'distance'],
    'feature_selection__k': [3, 5, 7, 9]
    },
    {'regressor': [HistGradientBoostingRegressor(random_state=42)],
    'regressor__max_iter': [100, 200, 500, 1000], 
    'regressor__learning_rate': [0.01, 0.1, 0.5],
    'feature_selection__k': [3, 5, 7, 9]
    }
]
```

```python
rand_search = RandomizedSearchCV(estimator=pipeline, param_distributions=par_grid, cv=5, n_iter=50, 
                                 scoring='r2', random_state=42)
```

```python
rand_search.fit(x_train_scaled, np.ravel(y_train))
```

```python
best_model = rand_search.best_estimator_
print(best_model)
```

```python
k_best = best_model.named_steps['feature_selection']
```

```python
best_features_mask = k_best.get_support()
best_features_mask
```

```python
best_features = X.columns[best_features_mask]
```

```python
best_features
```

```python
from sklearn.metrics import r2_score
y_pred = best_model.predict(x_test_scaled)
r2_score(y_test, y_pred)
```

```python
print(f'Accuracy: {round(best_model.score(x_test_scaled, y_test), 5)*100}%')
```

### Comparing the baseline performances of Models with the Best Model

```python
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
rfr = RandomForestRegressor()
lr = LinearRegression()
knr = KNeighborsRegressor()
hgbr = HistGradientBoostingRegressor()
```

```python
rfr.fit(x_train_scaled, y_train)
lr.fit(x_train_scaled, y_train)
knr.fit(x_train_scaled, y_train)
hgbr.fit(x_train_scaled, y_train)
```

```python
rfr_preds = rfr.predict(x_test_scaled)
lr_preds = lr.predict(x_test_scaled)
knr_preds = knr.predict(x_test_scaled)
hgbr_preds = hgbr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
print(f'RFR Accuracy: {round(r2_score(y_test, rfr_preds), 4) * 100} %')
print(f'LR Accuracy: {round(r2_score(y_test, lr_preds), 4) * 100} %')
print(f'KNR Accuracy: {round(r2_score(y_test, knr_preds), 4) * 100} %')
print(f'HGBR Accuracy: {round(r2_score(y_test, hgbr_preds), 4) * 100} %')
```

## Testing the best model

```python
x_best_cols = X.columns[best_features_mask]
X_best = X[x_best_cols]
```

```python
X_best.shape
```

```python
x_best_train, x_best_test, y_best_train, y_best_test = train_test_split(X_best, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import QuantileTransformer
scaler_best = QuantileTransformer()
x_best_train_scaled = scaler_best.fit_transform(x_best_train)
x_best_test_scaled = scaler_best.transform(x_best_test)
```

```python
best_fit_model = HistGradientBoostingRegressor(max_iter=200, random_state=42)
```

```python
best_fit_model.fit(x_best_train_scaled, y_best_train)
```

```python
best_model_preds = best_fit_model.predict(x_best_test_scaled)
```

```python
print(f'Best Model Accuracy: {round(r2_score(y_best_test, best_model_preds), 4) * 100} %')
```

## Plotting Results

```python
fig, axs = plt.subplots(2,2, figsize=(20,15))
fig.suptitle('Model Performance Report')

# rfr algo
axs[0][0].scatter(y_test, rfr_preds, color='red', label='RFR')
axs[0][0].set_title('Random Forest Regressor')
axs[0][0].legend()

# knr algo
axs[0][1].scatter(y_test, knr_preds, color='blue', label='KNR')
axs[0][1].set_title('KNeighbors Regressor')
axs[0][1].legend()

# ada_reg algo
axs[1][0].scatter(y_test, hgbr_preds, color='black', label='HGBR')
axs[1][0].set_title('HistGradientBoost Regressor')
axs[1][0].legend()

# ada_reg algo
axs[1][1].scatter(y_test, best_model_preds, color='green', label='best model-HGBR')
axs[1][1].set_title('Best Params - HistGradientBoost Regressor')
axs[1][1].legend()

plt.savefig('../reports/model_performance_report.png')
```
